```python
import pandas as pd
location = 'DC_Properties.csv'
df = pd.read_csv(location)
df.head()
```

```python
(df['PRICE']<100000) & (df['EYB'] > 2010)
```

```python
# Properties less than $100K and built before 2010
df[(df['PRICE']<100000) & (df['EYB'] > 2010)]
```

```python
# Properties less than $100K and built before 2010 and the mean number of rooms
df[(df['PRICE']<100000) & (df['EYB'] > 2010)]['ROOMS'].mean()
```

```python
# Properties less than $100K and built before 2010 and with 2 bathrooms
df[(df['PRICE']>1000000) & (df['EYB'] > 2010) & (df['BATHRM'] > 2)]
```

```python
df[(df['EYB'] < 1900) & (df['BATHRM'] == 2) & (df['PRICE'] < 200000)] #find properties built before 1900 with 2 bathrooms
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#make a scatterplot
sns.lmplot(x='X', y='Y', data=df[(df['EYB'] < 1900) & (df['BATHRM'] == 2) & (df['PRICE'] < 200000)])
```

```python
#change size of plot
plt.subplots(figsize=(30,15))

#boxplot showing stats for each characteristic
sns.boxplot(data=df[(df['EYB'] >= 2000) & (df['BATHRM'] == 1) & (df['PRICE'] < 200000)])
```

```python
#create a histogram of the distribution of PRICE
sns.distplot(df['EYB'])
```
